```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
income = fetch_ucirepo(id=20)

# data (as pandas dataframes)
a = income.data.features
b = income.data.targets

# metadata
print(income.metadata)

# variable information
print(income.variables)
```

```python
import pandas as pd
import numpy as np
```

```python
censusMerge = pd.concat([a, b  ], axis = 1)
censusMerge
```

```python
# prints the number of duplicates
censusMerge[censusMerge.duplicated()].shape[0]
```

```python
# dropping duplicates
censusMerge.drop_duplicates(inplace=True)
```

```python
# shows the sum of null data for every column
censusMerge.isnull().sum()
```

```python
censusMerge['native-country'].unique()
```

```python
changeNaN = {"?" : "Others"}

censusMerge.replace(changeNaN, inplace = True)
censusMerge.fillna('Others', inplace = True)
censusMerge
```

```python
censusMerge.replace('<=50K.', '<=50K', inplace=True)
censusMerge.replace('>50K.', '>50K', inplace=True)
```

```python
# Making Data Plots and Conducting Data Analysis
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(4, figsize = [10,20])

# marital-status column
censusMerge.groupby('marital-status').size().plot(kind='barh', ax = ax[0], color = ('midnightblue','lightseagreen'))
ax[0].set_title('Marital Status Population')
ax[0].set_xlabel('Number of People')


# relationship column
censusMerge.groupby('relationship').size().plot(kind='barh', ax = ax[1], color = ('midnightblue','lightseagreen'))
ax[1].set_title('Relationship Situation in the Community')
ax[1].set_xlabel('Number of People')


# sex column
censusMerge.groupby('sex').size().plot(kind='barh', ax = ax[2], color = ('midnightblue','lightseagreen'))
ax[2].set_title('Population\'s Sexual Orientations')
ax[2].set_xlabel('Number of People')


# race column
censusMerge.groupby('race').size().plot(kind='barh', ax = ax[3], color = ('midnightblue','lightseagreen'))
ax[3].set_title('Population\'s Race')
ax[3].set_xlabel('Number of People')


fig.tight_layout()
```

```python
# Using stacked bar graphs, plot each person's income according to work class
cen = censusMerge.sample(300, random_state=0)
cen.groupby('workclass')['age'].describe()
crosstab = pd.crosstab(censusMerge['workclass'], censusMerge['income'])
fig, ax = plt.subplots(figsize=(20,5))
crosstab.plot(kind='bar', stacked=True, ax=ax, color=['midnightblue', 'red'])
ax.set_title('Income of Individuals according to Workclass', fontsize=18, fontweight='bold')
ax.set_xticklabels(ax.get_xticklabels(), fontsize=16, rotation=90)
ax.set_yticklabels(ax.get_yticks(), fontsize=16)
ax.legend(fontsize=16)
```

```python
# utilizing barplots to plot each person's weekly hours by work class
cen.groupby('workclass')['hours-per-week'].describe()
fig, ax = plt.subplots(figsize=(15,5))
sns.barplot(x=cen['hours-per-week'], hue=cen['workclass'], ci=None, ax=ax)
ax.set_title('Weekly Hours of Individuals in Different Workclasses', fontsize=18, fontweight='bold')
ax.set_xticklabels(ax.get_xticklabels(), fontsize=10, rotation=90)
ax.set_yticklabels(ax.get_yticks(), fontsize=10)
ax.legend(fontsize=10)
sns.despine()
```
