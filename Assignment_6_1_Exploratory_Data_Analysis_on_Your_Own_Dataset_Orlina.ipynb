```python
import pandas as pd
filePath = '/content/League of Legends.csv'
data = pd.read_csv(filePath)
data
```

```python
# Identify the column names
column = list(data.columns)
print(column)
```

```python
# Identify the data types of the data
dataTypes = data.dtypes
print(dataTypes)
```

```python
# Display the total number of records
TotalRec = len(data)
print(TotalRec)
```

```python
# Display the first 20 records
first20 = data.head(20)
print(first20)
```

```python
# Display the last 20 records
last20 = data.tail(20)
print(last20)
```

```python
#Change the Player column to InGameName
data.rename(columns={'Player': 'InGameName'}, inplace=True)
data
```

```python
# Delete KDA column
data = data.drop('KDA', axis = 1)
data
```

```python
# New column for KDA ratio
def KDA(data):
  if data['Death'] == 0:
    return 0
  return (data['Kill'] + data['Assist']) / data['Death']

data['KDA'] = data.apply(KDA, axis=1)
data
```

```python
# Display the players who've won the match.
dragKill = data[data['Win/Lose'] == 'W']
total = len(dragKill)

print(dragKill)
print("\nNumber of players who have won the match:",total)
```

```python
# Display the players who've lost the match.
dragKill = data[data['Win/Lose'] == 'L']
total = len(dragKill)

print(dragKill)
print("\nNumber of players who have lost the match:",total)
```

```python
# Use numpy to get the average, median, and standard deviation of the KDA
import numpy as np
avgKDA = np.mean(data['KDA'])
medKDA = np.median(data['KDA'])
stdKDA = np.std(data['KDA'])

print("Average KDA:",avgKDA)
print("Median KDA:", medKDA)
print("Standard Deviation of KDA:", stdKDA)
```
