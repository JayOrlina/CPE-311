```python
import pandas as pd
filePath = '/content/LeaguePlayerStats.csv'
data = pd.read_csv(filePath)
data
```

```python
# Identify the column names
column = list(data.columns)
print(column)
```

```python
# Identify the data types of the data
dataTypes = data.dtypes
print(dataTypes)
```

```python
# Display the total number of records
TotalRec = len(data)
print(TotalRec)
```

```python
# Display the first 20 records
first20 = data.head(20)
print(first20)
```

```python
# Display the last 20 records
last20 = data.tail(20)
print(last20)
```

```python
#Change the summonerName column to InGameName
data.rename(columns={'summonerName': 'InGameName'}, inplace=True)
data
```

```python
# New column for KDA ratio
def KDA(data):
  return (data['kills'] + data['assists']) / data['deaths']

data['KDA'] = data.apply(KDA, axis=1)
data
```

```python
# Create a new dataframe that gathers data on dragon kills, then print the total number of players that killed dragons.
dragKill = data[data['dragonKills'] == 1]
total = len(dragKill)

print(dragKill)
print("\nNumber of players who have killed dragons:",total)
```

```python
# Create a new dataframe that gathers data on dragon kills, then print the total number of players who have not killed dragons.
dragKill = data[data['dragonKills'] == 0]
total = len(dragKill)

print(dragKill)
print("\nNumber of players who have killed dragons:",total)
```

```python
# Use numpy to get the average, median, and standard deviation of the KDA
import numpy as np
avgKDA = np.mean(data['KDA'])
medKDA = np.median(data['KDA'])
stdKDA = np.std(data['KDA'])

print("Average KDA:",avgKDA)
print("Median KDA:", medKDA)
print("Standard Deviation of KDA:", stdKDA)
```
